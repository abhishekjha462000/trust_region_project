# Model 4a

The following expression represents the **Model 4a function**:

$$
m(u_1, u_2, \ldots, u_{80}) = - g ^ T u \;+\;
\sum_{i=1}^{10} a_i \left( \sqrt{ (u_{2i-1})^2 + (1 + u_{2i})^2 } - 1 \right)^2
$$

$$
+ \sum_{i=1}^{9} a_{i+10} \left( \sqrt{ (1 + u_{2i+1})^2 + (1 + u_{2i+2})^2 } - \sqrt{2} \right)^2
$$

$$
+ \sum_{i=1}^{9} a_{i+19} \left( \sqrt{ (1 - u_{2i-1})^2 + (1 + u_{2i})^2 } - \sqrt{2} \right)^2
$$

$$
+ \sum_{i=1}^{30} a_{i+28} \left( \sqrt{ (1 + u_{2i+20} - u_{2i})^2 + (u_{2i+19} - u_{2i-1})^2 } - 1 \right)^2
$$

$$
+ \sum_{i=1}^{36} a_{i+58} \left( \sqrt{ \big(1 + u_{2i+1+2\lfloor \tfrac{i-1}{9} \rfloor} - u_{2i-1+2\lfloor \tfrac{i-1}{9} \rfloor}\big)^2 
+ \big(u_{2i+2+2\lfloor \tfrac{i-1}{9} \rfloor} - u_{2i+2\lfloor \tfrac{i-1}{9} \rfloor}\big)^2 } - 1 \right)^2
$$

$$
+ \sum_{i=1}^{27} a_{i+94} \left( \sqrt{ \big(1 + u_{2i+21+2\lfloor \tfrac{i-1}{9} \rfloor} - u_{2i-1+2\lfloor \tfrac{i-1}{9} \rfloor}\big)^2 
+ \big(1 + u_{2i+22+2\lfloor \tfrac{i-1}{9} \rfloor} - u_{2i+2\lfloor \tfrac{i-1}{9} \rfloor}\big)^2 } - \sqrt{2} \right)^2
$$

$$
+ \sum_{i=1}^{27} a_{i+121} \left( \sqrt{ \big(1 - u_{2i+19+2\lfloor \tfrac{i-1}{9} \rfloor} + u_{2i+1+2\lfloor \tfrac{i-1}{9} \rfloor}\big)^2 
+ \big(1 + u_{2i+20+2\lfloor \tfrac{i-1}{9} \rfloor} - u_{2i+2+2\lfloor \tfrac{i-1}{9} \rfloor}\big)^2 } - \sqrt{2} \right)^2
$$


```python
# Model 4A objective function

import numpy as np

def model_4a(u, g):
    """
        The function takes in 2 parameters namely g and u which are vectors from the 80th dimension euclidean space and calculates
        the objective function for that.
    """
     # you can write the dot multiplication this way as well g @ u
    m = -g @ u

    # Sum 1: i=1 to 10
    for i in range(1, 11):
        m += (np.sqrt(u[2 * i - 2]**2 + (1 + u[2 * i - 1]) ** 2) - 1) ** 2

    # Sum 2: i=1 to 9
    for i in range(1, 10):
        m += (np.sqrt((1 + u[2 * i]) ** 2 + (1 + u[2 * i + 1]) ** 2) - (2 ** (1 / 2))) ** 2

    # Sum 3: i=1 to 9
    for i in range(1, 10):
        m += (np.sqrt((1 - u[2 * i - 2]) ** 2 + (1 + u[2 * i - 1]) ** 2) - (2 ** (1 / 2))) ** 2

    # Sum 4: i=1 to 30
    for i in range(1, 31):
        m += (np.sqrt((1 + u[2 * i + 19] - u[2 * i - 1]) ** 2 + 
                         (u[2 * i + 18] - u[2 * i - 2]) ** 2) - 1) ** 2

    # Sum 5: i=1 to 36
    for i in range(1, 37):
        floor_val = (i - 1) // 9
        m += (np.sqrt((1 + u[2 * i + 2 *  floor_val] - u[2 * i - 2 + 2 * floor_val]) ** 2 + 
                         (u[2 * i + 1 + 2 * floor_val] - u[2 * i - 1 + 2 * floor_val]) ** 2) - 1) ** 2

    # Sum 6: i=1 to 27
    for i in range(1, 28):
        floor_val = (i - 1) // 9
        m += (np.sqrt((1 + u[2 * i + 20 + 2 * floor_val] - u[2 * i - 2 + 2 * floor_val]) ** 2 + 
                         (1 + u[2 * i + 21 + 2 * floor_val] - u[2 * i - 1 + 2 * floor_val]) ** 2) - (2 ** (1 / 2)))**2

    # Sum 7: i=1 to 27
    for i in range(1, 28):
        floor_val = (i - 1) // 9
        m += (np.sqrt((1 - u[2 * i + 18 + 2 * floor_val] + u[2 * i + 2 * floor_val]) ** 2 + 
                         (1 + u[2 * i + 19 + 2 * floor_val] - u[2 * i + 1 + 2 * floor_val]) ** 2) - (2 ** (1 / 2)))**2

    return m
```

```python
# Gradient function
def calculate_gradient(u, g):
    grad = -g.copy()  # derivative of -g@u

    # Sum 1
    for i in range(1, 11):
        x, y = u[2*i-2], u[2*i-1]
        r = np.sqrt(x**2 + (1+y)**2)
        coeff = 2*(r-1)/r
        grad[2*i-2] += coeff*x
        grad[2*i-1] += coeff*(1+y)

    # Sum 2
    for i in range(1, 10):
        x, y = u[2*i], u[2*i+1]
        r = np.sqrt((1+x)**2 + (1+y)**2)
        coeff = 2*(r-np.sqrt(2))/r
        grad[2*i]   += coeff*(1+x)
        grad[2*i+1] += coeff*(1+y)

    # Sum 3
    for i in range(1, 10):
        x, y = u[2*i-2], u[2*i-1]
        r = np.sqrt((1-x)**2 + (1+y)**2)
        coeff = 2*(r-np.sqrt(2))/r
        grad[2*i-2] += coeff*(x-1)
        grad[2*i-1] += coeff*(1+y)

    # Sum 4
    for i in range(1, 31):
        x1 = 1 + u[2*i+19] - u[2*i-1]
        x2 = u[2*i+18] - u[2*i-2]
        r = np.sqrt(x1**2 + x2**2)
        coeff = 2*(r-1)/r
        grad[2*i+19] += coeff*x1
        grad[2*i-1]  -= coeff*x1
        grad[2*i+18] += coeff*x2
        grad[2*i-2]  -= coeff*x2

    # Sum 5
    for i in range(1, 37):
        k = 2*((i-1)//9)
        x1 = 1 + u[2*i+k] - u[2*i-2+k]
        x2 = u[2*i+1+k] - u[2*i-1+k]
        r = np.sqrt(x1**2 + x2**2)
        coeff = 2*(r-1)/r
        grad[2*i+k]     += coeff*x1
        grad[2*i-2+k]   -= coeff*x1
        grad[2*i+1+k]   += coeff*x2
        grad[2*i-1+k]   -= coeff*x2

    # Sum 6
    for i in range(1, 28):
        k = 2*((i-1)//9)
        x1 = 1 + u[2*i+20+k] - u[2*i-2+k]
        x2 = 1 + u[2*i+21+k] - u[2*i-1+k]
        r = np.sqrt(x1**2 + x2**2)
        coeff = 2*(r-np.sqrt(2))/r
        grad[2*i+20+k] += coeff*x1
        grad[2*i-2+k]  -= coeff*x1
        grad[2*i+21+k] += coeff*x2
        grad[2*i-1+k]  -= coeff*x2

    # Sum 7
    for i in range(1, 28):
        k = 2*((i-1)//9)
        x1 = 1 - u[2*i+18+k] + u[2*i+k]
        x2 = 1 + u[2*i+19+k] - u[2*i+1+k]
        r = np.sqrt(x1**2 + x2**2)
        coeff = 2*(r-np.sqrt(2))/r
        grad[2*i+18+k] -= coeff*x1
        grad[2*i+k]    += coeff*x1
        grad[2*i+19+k] += coeff*x2
        grad[2*i+1+k]  -= coeff*x2

    return grad
```

```python
grad_val = calculate_gradient(u, g)
print("\nGradient vector:\n", np.round(grad_val, 4))
```

# Hessian of the model 4A

```python
import re

def sympy_to_python_converter(expression):
    expr = expression  # Paste your full expression
    # Replace sqrt(...) with (... )**(1/2)
    expr = re.sub(r'sqrt\((.*?)\)', r'(\1)**(1/2)', expr)

    # Replace aX  a[X-1]
    expr = re.sub(r'\ba(\d+)\b', lambda m: f'a[{int(m.group(1))-1}]', expr)

    # Replace uX  u[X-1]
    expr = re.sub(r'\bu(\d+)\b', lambda m: f'u[{int(m.group(1))-1}]', expr)

    # Replace gX  g[X-1]
    expr = re.sub(r'\bg(\d+)\b', lambda m: f'g[{int(m.group(1))-1}]', expr)

    return expr
```

```python
def hess_single(A, B, c):
    r2 = A*A + B*B
    r = np.sqrt(r2)

    if r < 1e-14:     # avoid singularity
        return np.zeros((2,2))

    f = r - c
    A2 = A*A
    B2 = B*B

    # Common term
    t = (2 - 2*f/r) / r2

    H = np.zeros((2,2))
    H[0,0] = 2*f*A2/(r2*r) + t*B2
    H[1,1] = 2*f*B2/(r2*r) + t*A2
    H[0,1] = H[1,0] = -A*B*(2*f/r2/r + t/r)

    return H


def hessian_model4a(u, g=None):
    """
    Computes the full 80x80 Hessian matrix of Model 4a,
    with all a_i = 1.
    """
    H = np.zeros((80,80))

    # --- TERM GROUP 1: -g^T u has ZERO Hessian ---

    # --- TERM GROUP 2 ---
    for i in range(1, 11):
        A = u[2*i - 2]
        B = 1 + u[2*i - 1]

        Hloc = hess_single(A, B, 1)

        idx = [2*i-2, 2*i-1]
        H[np.ix_(idx, idx)] += Hloc

    # --- TERM GROUP 3 ---
    for i in range(1, 10):
        A = 1 + u[2*i]
        B = 1 + u[2*i + 1]

        Hloc = hess_single(A, B, np.sqrt(2))

        idx = [2*i, 2*i+1]
        H[np.ix_(idx, idx)] += Hloc

    # --- TERM GROUP 4 ---
    for i in range(1, 10):
        A = 1 - u[2*i - 2]
        B = 1 + u[2*i - 1]

        # Derivatives pick up -1 for A w.r.t that index:
        signs = np.array([-1, +1])

        Hloc = hess_single(A, B, np.sqrt(2))

        idx = [2*i-2, 2*i-1]
        H[np.ix_(idx, idx)] += (signs[:,None] * signs[None,:]) * Hloc

    # --- TERM GROUP 5 ---
    for i in range(1, 31):
        A = 1 + u[2*i+19] - u[2*i - 1]
        B = u[2*i+18] - u[2*i - 2]

        signs = np.array([-1, +1])   # corresponding to derivative direction

        Hloc = hess_single(A, B, 1)

        idx = [2*i - 1, 2*i + 19]   # A depends on these two
        H[np.ix_(idx, idx)] += np.array([[1, -1],
                                         [-1, 1]]) * Hloc[0,0]  # manually place blocks

        idx2 = [2*i - 2, 2*i + 18]  # B depends on these two
        H[np.ix_(idx2, idx2)] += np.array([[1, -1],
                                           [-1, 1]]) * Hloc[1,1]

        # Cross terms
        H[np.ix_(idx, idx2)] += Hloc[0,1] * np.array([[1, -1],
                                                      [-1, 1]])
        H[np.ix_(idx2, idx)] += Hloc[1,0] * np.array([[1, -1],
                                                      [-1, 1]])

    # --- TERM GROUPS 6,7,8 ---
    for group in [6,7,8]:
        for i in range(1, 28 if group>5 else 37):
            k = 2*((i-1)//9)

            if group == 6:
                A = 1 + u[2*i + k] - u[2*i + k - 2]
                B = u[2*i + k + 1] - u[2*i + k - 1]
                c = 1

                Ia = [2*i + k, 2*i + k - 2]
                Ib = [2*i + k + 1, 2*i + k - 1]

                Sa = np.array([+1, -1])
                Sb = np.array([+1, -1])

            elif group == 7:
                A = 1 + u[2*i + k + 20] - u[2*i + k - 2]
                B = 1 + u[2*i + k + 21] - u[2*i + k - 1]
                c = np.sqrt(2)

                Ia = [2*i + k + 20, 2*i + k - 2]
                Ib = [2*i + k + 21, 2*i + k - 1]

                Sa = np.array([+1, -1])
                Sb = np.array([+1, -1])

            else:  # group 8
                A = 1 - u[2*i + k + 18] + u[2*i + k]
                B = 1 + u[2*i + k + 19] - u[2*i + k + 1]
                c = np.sqrt(2)

                Ia = [2*i + k, 2*i + k + 18]
                Ib = [2*i + k + 1, 2*i + k + 19]

                Sa = np.array([+1, -1])
                Sb = np.array([-1, -1])

            Hloc = hess_single(A, B, c)

            # Block placements
            H[np.ix_(Ia, Ia)] += (Sa[:,None]*Sa[None,:])*Hloc[0,0]
            H[np.ix_(Ib, Ib)] += (Sb[:,None]*Sb[None,:])*Hloc[1,1]
            H[np.ix_(Ia, Ib)] += (Sa[:,None]*Sb[None,:])*Hloc[0,1]
            H[np.ix_(Ib, Ia)] += (Sb[:,None]*Sa[None,:])*Hloc[1,0]

    return H
```
